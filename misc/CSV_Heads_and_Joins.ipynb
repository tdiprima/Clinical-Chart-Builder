# === Simplified Join Examples ===

```python
# Example 1: Join visit_occurrence to concept to decode visit_concept_id
visit_with_concept = dfs["visit_occurrence"].merge(
    dfs["concept"], left_on="visit_concept_id", right_on="concept_id", how="left"
)
visit_with_concept[["person_id", "visit_start_date", "visit_concept_id", "concept_name"]].head()
```

```python
# Example 2: Join condition_occurrence to concept to decode condition_concept_id
condition_with_concept = dfs["condition_occurrence"].merge(
    dfs["concept"], left_on="condition_concept_id", right_on="concept_id", how="left"
)
condition_with_concept[["person_id", "condition_start_date", "condition_concept_id", "concept_name"]].head()
```

```python
# Example 3: Join condition_occurrence to visit_occurrence
condition_with_visit = dfs["condition_occurrence"].merge(
    dfs["visit_occurrence"], on="visit_occurrence_id", how="left"
)

# Use renamed columns due to merge conflict
condition_with_visit[[
    "person_id_x",  # from condition_occurrence
    "visit_occurrence_id",
    "condition_concept_id",
    "visit_start_date"
]].head()
```

You just need to add one more merge, this time to bring in the concept_name for the condition_concept_id. Here's how to do it, chained onto the last merge:

```python
# Step 1: Join condition_occurrence to visit_occurrence
condition_with_visit = dfs["condition_occurrence"].merge(
    dfs["visit_occurrence"], on="visit_occurrence_id", how="left"
)

# Step 2: Join to concept on condition_concept_id to get concept_name
condition_with_visit_and_concept = condition_with_visit.merge(
    dfs["concept"], left_on="condition_concept_id", right_on="concept_id", how="left"
)

# Step 3: Show relevant columns
condition_with_visit_and_concept[[
    "person_id_x",          # from condition_occurrence
    "visit_occurrence_id",
    "condition_concept_id",
    "visit_start_date",
    "concept_name"          # human-readable condition name
]].head()
```

If you want to clean up person_id_x into just person_id for readability:

```python
condition_with_visit_and_concept.rename(columns={"person_id_x": "person_id"}, inplace=True)
condition_with_visit_and_concept[["person_id", "visit_occurrence_id", "condition_concept_id", "visit_start_date", "concept_name"]].head()
```

Boom  now you've got conditions tied to visits and decoded into real names 

You've already joined `condition_occurrence  visit_occurrence  concept` for `condition_concept_id`.

Now we'll **also decode** `visit_concept_id` (which lives in `visit_occurrence`) using another join to `concept`.

Here's the full version with **both** concept names decoded:


```python
# Step 1: Join condition_occurrence to visit_occurrence
condition_with_visit = dfs["condition_occurrence"].merge(
    dfs["visit_occurrence"], on="visit_occurrence_id", how="left"
)

# Step 2: Join to concept to decode condition_concept_id
with_condition_name = condition_with_visit.merge(
    dfs["concept"], left_on="condition_concept_id", right_on="concept_id", how="left"
).rename(columns={"concept_name": "condition_name"})

# Step 3: Join again to decode visit_concept_id
fully_decoded = with_condition_name.merge(
    dfs["concept"], left_on="visit_concept_id", right_on="concept_id", how="left"
).rename(columns={"concept_name": "visit_type_name"})

# Step 4: Optional cleanup
fully_decoded = fully_decoded.rename(columns={"person_id_x": "person_id"})

# Step 5: Show the final merged view
fully_decoded[[
    "person_id",
    "visit_occurrence_id",
    "condition_concept_id",
    "condition_name",
    "visit_concept_id",
    "visit_type_name",
    "visit_start_date"
]].head()
```

This gives you:
- `condition_name`: like "Hypertension"
- `visit_type_name`: like "Inpatient Visit"

##  Time Series: Conditions per Month

```python
# Make sure visit_start_date is datetime
fully_decoded["visit_start_date"] = pd.to_datetime(fully_decoded["visit_start_date"])

# Create year-month column
fully_decoded["year_month"] = fully_decoded["visit_start_date"].dt.to_period("M")

# Group by time and count conditions
ts = fully_decoded.groupby("year_month")["condition_concept_id"].count().reset_index()
ts.columns = ["month", "condition_count"]

# Display it
ts.sort_values("month").head(10)
```

```python
# === TIME SERIES: Condition counts by month ===
fully_decoded["year_month"] = fully_decoded["visit_start_date"].dt.to_period("M").dt.to_timestamp()
ts = fully_decoded.groupby("year_month")["condition_concept_id"].count().reset_index()
ts.columns = ["month", "condition_count"]

# Plot time series
plt.figure(figsize=(10,5))
plt.plot(ts["month"], ts["condition_count"], marker="o")
plt.title("Conditions Diagnosed Over Time")
plt.xlabel("Month")
plt.ylabel("Condition Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

##  Grouped Count: Top 10 Most Common Conditions

```python
# Group by condition name and count
common_conditions = fully_decoded.groupby("condition_name")["person_id"].count().reset_index()
common_conditions.columns = ["condition_name", "count"]

# Sort and show top 10
common_conditions.sort_values("count", ascending=False).head(10)
```

```python
# === GROUPED COUNT: Most common conditions ===
top_conditions = common_conditions.sort_values("count", ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,5))
plt.barh(top_conditions["condition_name"][::-1], top_conditions["count"][::-1])
plt.title("Top 10 Most Common Conditions")
plt.xlabel("Count")
plt.tight_layout()
plt.show()
```
